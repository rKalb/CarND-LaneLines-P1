Below are some helper functions to help get you started. They should look familiar from the lesson!

```python
import math

# Hold onto the last slope, frame by frame
last_slope_L = 0.0
last_slope_R = 0.0
last_b_L = 0.0
last_b_R = 0.0

def print_all_images(list_of_images):
    for each_image_file in list_of_images:
        each_image_file = '{0}{1}'.format(image_path,each_image_file) # Add the path
        image = mpimg.imread(each_image_file)
        plt.figure()
        plt.imshow(image)
        print (each_image_file)
    return

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, Gaussian_kernel):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (Gaussian_kernel, Gaussian_kernel), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    # K: Note this assumes you're using ff for your bitwise mask.  If not, the results will be...unexpected
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, color=[255, 0, 0], thickness=2):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    for line in lines:
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)
            
def make_line(slope, b, Ymax, Ymid):
    # Calculate the average slopes    
    y1 = Ymax
    x1 = int((y1 - b) // slope)

    # Calculate the x1, y1, x2, y2 for each lane line
    # Use the value that is smallest between L and R
    y2 = Ymid
    x2 = int((y2 - b) // slope)

    return ((x1, y1), (x2,y2))

def video_frames(stuff):
    clip1 = VideoFileClip("solidYellowLeft.mp4")
    myframe = clip1.get_frame(25.2)
    newImage = process_image(myframe)

def make_LR(lines, line_image, min_slope = 0.4):
    yellow = (255,255,0)
    white = (255,255,255)
    red = (255,0,0)
    green = (0,255,0)

    L_num = 0
    L_slope_list = []
    L_b = 0
    L_minY = line_image.shape[0]
    
    R_num = 0
    R_slope_list = []
    R_b = 0
    R_minY = line_image.shape[0]
    
    if lines == []:
        raise Exception( "No Hough lines were found")
        print ("No Hough lines found!!")
        return None
    
    # Iterate over the output "lines" and draw lines on a blank image
    for line in lines:
        for x1,y1,x2,y2 in line:
            slope = (y2 - y1) / (x2 - x1)
            # Find the left lane lines
            if slope < -min_slope:
                L_minY = min(L_minY,min(y1,y2)) # The is the vanishing point
                L_slope_list.append(slope) # Save these to take the median
                L_b += (y2 - slope * x2)  # y = mx + b
                L_num += 1
                # print ("Slope: {} X: {} Y:{} ".format(slope, x1 - x2, y1 - y2))
                # cv2.line(line_image,(x1,y1),(x2,y2),green,10)
            elif slope > min_slope:
                R_minY = min(R_minY,min(y1,y2))
                R_slope_list.append(slope)
                R_b += (y2 - slope * x2) # y = mx + b
                R_num += 1
                # cv2.line(line_image,(x1,y1),(x2,y2),green,10)
               
    if L_num == 0 or R_num == 0:
        print (L_slope_list)
        print("No lines!!")
        print (R_slope_list)
        plt.figure()
        plt.imshow(line_image, cmap="gray")
        return line_image
                                
    # Draw the left line in yellow
    # slope = np.median(np.array(L_slope_list))
    slope = np.mean(np.array(L_slope_list))
    b = L_b//L_num
    Ymax = line_image.shape[0]
    Ymid = min(L_minY, R_minY)
    line = make_line(slope, b, Ymax, Ymid)
    cv2.line(line_image,line[0],line[1],red,5)
    # Draw the right line in white
    # slope = np.median(np.array(R_slope_list))
    slope = np.mean(np.array(R_slope_list))
    b = R_b//R_num
    line = make_line(slope, b, Ymax, Ymid)
    cv2.line(line_image,line[0],line[1],red,5)
    return line_image

#  = rho is the radius/distance of the point from the origin
#  = theta is the angle of the point from the origin
# A line in the image corresponds to a point in / Hough space
def hough_lines(Canny_img, , , threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(Canny_img, , , threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((Canny_img.shape[0], Canny_img.shape[1], 3), dtype=np.uint8)
    # draw_lines(line_img, lines, color=[0,255,0],thickness=10)

    return make_LR(lines, line_img)

# Python 3 has support for cool math symbols.
# = alpha
# = beta
# = lambda
def weighted_img(img, initial_img, =0.8, =1., =0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img *  + img *  + 
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, , img, , )
```

```python
def process_image(each_image):
    #Convert to grayscale
    gray = grayscale(each_image)
    # Blur the gray image
    blur_gray = cv2.GaussianBlur(gray,(Gaussian_kernel, Gaussian_kernel),0)
    # Then find the edges
    edges = cv2.Canny(blur_gray, low_threshold, high_threshold)
    
    # Mask a region of interest
    ignore_mask_color = 255
    xsize = each_image.shape[1]
    ysize = each_image.shape[0]
    vertices = np.array([[(xsize/20, ysize),(xsize / 2.2, ysize/1.7), (xsize / 1.8, ysize/1.7), (xsize - xsize/25, ysize)]], dtype=np.int32)

    masked_edges = region_of_interest(edges, vertices)

    # Run Hough on edge detected image
    line_image = hough_lines(masked_edges, rho, theta, threshold, min_line_length, max_line_gap)
    combined_img = weighted_img(line_image, each_image, 1.0, 0.5)

    return combined_img
```
